```python
#Q1
cont = 0
def hannoi(n, orig, dest, aux):
    global cont
    if n == 1:
        cont +=1
        print(f"mover o disco {n} da torre {orig} para a torre {dest}") 
        return
    hannoi(n-1, orig, aux, dest)
    cont += 1
    print(f"mover o disco {n} da torre {orig} para a torre {dest}") 
    hannoi(n-1, aux, dest, orig)
n = int(input("Digite o nº de discos:\n"))
hannoi(n, 1, 3, 2)
print(f"Nº de movimentos = {cont}")
```

```python
#Q2
def fact(num):
    if num == 0 or num == 1:
        return 1
    return num * fact(num - 1)

def comb(passos, duplos):
    div1 = fact(passos)
    div2 = fact(duplos) * fact(passos-duplos)
    return int(div1/div2)

n = int(input("Digite o numero de degraus: (999 max)\n"))
cont = 1

dup = n//2
if n > 1:
    for i in range(1, dup+1):
        steps = n - i
        cont += comb(steps, i)

print(f'Existem {cont} maneira(s) de subir uma escada com {n} degrau(s)')
```

### **Q4:**

**($\Rightarrow$)**

Se $g(n) = O(f(n))$, então $\exists c_1, N_1$ tais que:
$$
    g(n) \leq c_1f(n), \quad \forall n \geq N_1
$$
$$
    \frac{1}{c_1}g(n) = O(f(n))
$$
Então, $c_2 = \frac{1}{c_1}$ e $N_2 = N_1 \Rightarrow f(n)= \Omega (g(n))$

---
**($\Leftarrow$)**

Analogamente, se $f(n) = \Omega(g(n))$, então $\exists c_2, N_2$ tais que:
$$
    f(n) \leq c_2g(n), \quad \forall n \geq N_2
$$
$$
    \frac{1}{c_2}f(n) = \Omega(g(n))
$$
Então, $c_1 = \frac{1}{c_2}$ e $N_1 = N_2 \Rightarrow f(n) = \Omega(f(n))$
