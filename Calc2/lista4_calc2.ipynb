```sage
#Questão 6 Apostol 8.5

%matplotlib inline

@interact
def plot_campo(
    m=slider(-2, 2, step_size=0.1, default=1, label='m')
):
    i = 2 #intervalo regular

    #z constante em função de m:
    constante = (1 - m^2)/(1 + m^2)

    f(x, y) = (x^2 - y^2)/(x^2 + y^2) if (x, y) != (0, 0) else None

    var('x y z t')

    campo = implicit_plot3d(z == f, (x, -i, i), (y, -i, i), (z, -i, i),
                            color = 'purple',
                            opacity= 0.5,
                            axes = True
                            )
    
    reta1 = parametric_plot3d(
        (t, m*t, f(t, m*t)), # A curva (x, y, z) em função de t
        (t, -i, i),
        color='blue',
        thickness=4
    )

    reta2 = parametric_plot3d(
        (t, (-m)*t, f(t, (-m)*t)), # A curva (x, y, z) em função de t
        (t, -i, i),
        color='red',
        thickness=4
    )
    
    plot = campo + reta1 + reta2

    plot.show()

    print(f'Para a reta y = {m:.1f}*x, o valor de f(x, y) é constante e igual a: {constante:.4f}')
    print('As retas são a representação geométrica do limite da função na origem')
```

```sage
#Questão 4 apostol 8.9
var('x, y')

f4(x, y) = x^2 + y^2 * sin(x*y)

print("Questão 4")
print(f"f(x,y) = {f4(x,y)}")
print(f"df/dx = {df4_dx}")
print(f"df/dy = {df4_dy}")

df4_dx = diff(f4, x)
df4_dy = diff(f4, y)

plot_f4_surface = plot3d(f4, (x, -3, 3), (y, -3, 3),
                         cmap='hot',
                         opacity=0.8,
                         axes_labels=['x', 'y', 'f(x,y)'])
plot_f4_surface.show()

grad_f4 = vector([df4_dx, df4_dy])

plot_grad_f4 = plot_vector_field(grad_f4, (x, -3, 3), (y, -3, 3),
                                 color='red',
                                 plot_points=20,
                                 aspect_ratio=1)

plot_contours_f4 = contour_plot(f4, (x, -3, 3), (y, -3, 3),
                                 cmap='hot',
                                 contours=15,
                                 fill=False)

combined_plot_f4 = plot_grad_f4 + plot_contours_f4
combined_plot_f4.show()
```

```sage
#Questão 1 apostol 8.14

var('x, y, z')

f_c(x, y, z) = x^2 * y^3 * z^4
dfc_dx = diff(f_c, x)
dfc_dy = diff(f_c, y)
dfc_dz = diff(f_c, z)

print("--- Item (c) ---")
print(f"f(x,y,z) = {f_c(x,y,z)}")
print(f"df/dx = {dfc_dx}")
print(f"df/dy = {dfc_dy}")
print(f"df/dz = {dfc_dz}")
print("------------------\n")


f_d(x, y, z) = x^2 - y^2 + 2*z^2
dfd_dx = diff(f_d, x)
dfd_dy = diff(f_d, y)
dfd_dz = diff(f_d, z)

print("--- Item (d) ---")
print(f"f(x,y,z) = {f_d(x,y,z)}")
print(f"df/dx = {dfd_dx}")
print(f"df/dy = {dfd_dy}")
print(f"df/dz = {dfd_dz}")
print("------------------\n")

f_e(x, y, z) = log(x^2 + 2*y^2 - 3*z^2)
dfe_dx = diff(f_e, x)
dfe_dy = diff(f_e, y)
dfe_dz = diff(f_e, z)

print("--- Item (e) ---")
print(f"f(x,y,z) = {f_e(x,y,z)}")
print(f"df/dx = {dfe_dx}")
print(f"df/dy = {dfe_dy}")
print(f"df/dz = {dfe_dz}")
print("------------------\n")

f_f(x, y, z) = x^(y^z)
dff_dx = diff(f_f, x)
dff_dy = diff(f_f, y)
dff_dz = diff(f_f, z)

print("--- Item (f) ---")
print(f"f(x,y,z) = {f_f(x,y,z)}")
print(f"df/dx = {dff_dx}")
print(f"df/dy = {dff_dy}")
print(f"df/dz = {dff_dz}")
print("------------------\n")
```

```sage
#Valor médio
import matplotlib.cm as cm

var('x, y, u')

f(x, y) = x^2 + y^2

P_x, P_y = -1.5, -1.0
Q_x, Q_y = 1.5, 1.0
P = vector([P_x, P_y])
Q = vector([Q_x, Q_y])

f_P = float(f(P[0], P[1]))
f_Q = float(f(Q[0], Q[1]))
lado_esquerdo = f_Q - f_P

def grad_f(x_val, y_val):
    return vector([2*x_val, 2*y_val])

def produto_interno_t(t):
    M_temp = P + t*(Q - P)
    grad_M = grad_f(M_temp[0], M_temp[1])
    return float(grad_M.dot_product(Q - P))

try:
    t_M = find_root(lambda t: produto_interno_t(t) - lado_esquerdo, 0, 1)
except:
    t_M = 0.5

M_dominio = P + t_M*(Q - P)

plt = contour_plot(f, (x, -2, 2), (y, -2, 2), cmap= cm.plasma, fill=False)
plt += line([P, Q], color='cyan', thickness=2, linestyle='-')
plt += point(P, color='magenta', size=40)
plt += point(Q, color='orange', size=40)
plt += point(M_dominio, color='white', size=30, zorder=10)

show(plt)

P_superficie = vector([P_x, P_y, f_P])
Q_superficie = vector([Q_x, Q_y, f_Q])
M_x = float(M_dominio[0])
M_y = float(M_dominio[1])
M_superficie = vector([M_x, M_y, float(f(M_x, M_y))])

superficie_3d = plot3d(f(x, y), (x, -2, 2), (y, -2, 2), 
                      color='lightgray', opacity=0.8)

superficie_3d += point3d(P_superficie, color='magenta', size=80)
superficie_3d += point3d(Q_superficie, color='orange', size=80)
superficie_3d += point3d(M_superficie, color='black', size=80)

curva_3d = parametric_plot3d(
    (P[0] + u*(Q[0]-P[0]), P[1] + u*(Q[1]-P[1]), 
     f(P[0] + u*(Q[0]-P[0]), P[1] + u*(Q[1]-P[1]))),
    (u, 0, 1), color='red', thickness=4
)

secante_3d = line([P_superficie, Q_superficie], color='cyan', thickness=3, linestyle='--')

show(superficie_3d + curva_3d + secante_3d)
```
