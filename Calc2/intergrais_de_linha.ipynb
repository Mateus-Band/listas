```python
var('x, y, t')

def integral_de_linha(alpha, f, a, b, t1, t2):
    dalpha = diff(alpha(t), t)
    plot = parametric_plot(alpha(t), (t, a, b), color='blue', legend_label='alpha(t)')

    y_t(t) = alpha(t)[1]
    y_prime = diff(y_t(t), t)
    critical_points = solve(y_prime == 0, t)
    valid_critical_points = [sol.rhs() for sol in critical_points if a < sol.rhs() < b]

    values_to_check = [y_t(a), y_t(b)] + [y_t(cp) for cp in valid_critical_points]

    y_min_val = min(values_to_check)
    y_max_val = max(values_to_check)

    margem = (y_max_val - y_min_val) * 0.1 + 0.1
    y_min = y_min_val - margem
    y_max = y_max_val + margem


    vector_field = plot_vector_field(f(x, y), (x, a, b), (y, y_min, y_max), color='red')
    for ti in [t1, t2]:
        ponto = alpha(ti)
        d_alpha = diff(alpha(t), t).subs(t=ti)
        f_alpha = f(ponto[0], ponto[1])
        
        plot += point(ponto, size=30, color='black')
        plot += arrow(ponto, ponto + d_alpha, color='green', width=1)
        plot += arrow(ponto, ponto + f_alpha, color='orange', width=1)

    plot += vector_field
    plot.show()

    produto_interno = f(alpha(t)[0], alpha(t)[1]).dot_product(dalpha)
    integral = integrate(produto_interno, (t, a, b))

    print("O valor da integral de linha Ã©:", integral)
    
```

```python
#Q1

a = -1
b = 1

alpha(t) = (t, t^2)
f(x, y) = (x^2 - 2*x*y, y^2 - 2*x*y)

integral_de_linha(alpha, f, a, b, -0.5, -0.25)

```
