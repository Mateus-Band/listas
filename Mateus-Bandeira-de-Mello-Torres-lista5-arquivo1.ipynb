```python
#('##' são testes)

#Classe questão 1
class Aluno:
    def __init__(self, nome, notas):
        self.nome = nome
        self.notas = notas
    
    def media(self):
        notas_ordenadas = sorted(self.notas)
        notas_validas = notas_ordenadas[2:]
        return sum(notas_validas) / len(notas_validas)

    def aprovado(self):
        return self.media() >= 6.0

    def __repr__(self):
        return f'Aluno(nome={self.nome}, media={self.media():.1f}, aprovado={self.aprovado()})'
    
    def media_total(self):
        return sum(self.notas)/ len(self.notas)

alunos = []

import csv
#caminho para o arquivo no meu notebook
with open('/home/mateus-band/Code/padrao/codes/txts/alunos_notas.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader) #Pular o cabeçalho
    for row in reader:
        nome = row[0]
        notas = list(map(int, row[1:]))
        alunos.append(Aluno(nome, notas))


'''for a in alunos:
    print(repr(a))'''

#item a) questão 1
medias1 = list(map(Aluno.media, alunos))
print([f'{m:.1f}' for m in medias1], '\n')

#item b)
aprovados = list(filter(Aluno.aprovado, alunos))
print(aprovados, '\n')

#item c)
from functools import reduce
soma_medias = reduce(lambda x, y: x + y, map(Aluno.media, alunos))
media_geral = soma_medias / len(alunos)
print(f'Média geral da turma: {media_geral:.1f}\n')

#item a) (compreensão de listas)
medias2 = [a.media() for a in alunos]
print([f'{m:.1f}' for m in medias2], '\n')

#item b) (compreensão de listas)
aprovados2 = [a for a in alunos if a.aprovado()]
print(aprovados2)
```

__Código gerador de mapas para as questões 2, 3 e 4__

```python
import random


def generateMap(m: int, n: int, ground_water_ration = .2, water = '-', ground = '*' ):
    r = int(m*n*ground_water_ration+.5)
    newMap = [[water]*m for _ in range(n)]
    coord = [(i, j) for i in range(n) for j in range(m)]
    random.shuffle(coord)
    for i, j in coord[:r]:
        newMap[i][j] = ground
    return newMap

def save_map(map_s, path = 'new_map.txt'):
    with open(path, 'wt') as f:
        for row in map_s:
            f.write("".join(map(str, row)))
            f.write('\n')

def print_map(map_p, ground = '*'):
    RESET = "\033[0m"
    AZULCLARO = '\033[94m'
    MARROM = '\033[33m'

    for row in map_p:
        for char in row:
            if char == ground:
                print(f'{MARROM}{char}{RESET}', end='')
            else:
                print(f'{AZULCLARO}{char}{RESET}', end='')
        print()

if __name__ == '__main__':
    random.seed(10012)

    m1 = generateMap(50, 10, 0.1)
    print_map(m1)

    m2 = generateMap(100, 120)
    save_map(m2, 'map_ilhas.txt')
    

    
```

```python
#Nova função para retornar se existe lago(s) na matriz e quantos:
def conta_lagos(matriz, agua='0'):
    if not matriz:
        return 0
    
    m = len(matriz)
    n = len(matriz[0])

    visitados = set()
    cont = 0

    def dfs(i, j):
        m = len(matriz)
        n = len(matriz[0]) if m else 0
        stack = [(i, j)]
        while stack:
            i, j = stack.pop()
            if not (0 <= i < m and 0 <= j < n):
                continue
            if (i, j) in visitados:
                continue
            if matriz[i][j] != agua:
                continue
            visitados.add((i, j))
            for di, dj in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
                stack.append((i + di, j + dj))

    #primeira rodagem para retirar o que for oceano:
    for i in range(m):
        if (i, 0) not in visitados and matriz[i][0] == agua:
            dfs(i, 0)
        if (i, n-1) not in visitados and matriz[i][n-1] == agua:
            dfs(i, n-1)
    for j in range(n):
        if (0, j) not in visitados and matriz[0][j] == agua:
            dfs(0, j)
        if (m-1, j) not in visitados and matriz[m-1][j] == agua:
            dfs(m-1, j)


    for i in range(m):
        for j in range(n):
            if (i, j) not in visitados and matriz[i][j] == agua:
                dfs(i, j)
                cont += 1


    return cont

with open('map_lagos.txt', 'r') as f:
    matriz_lagos = [list(line.strip()) for line in f if line.strip()]

print_map(matriz)

agua = '-'

contagem_lagos = conta_lagos(matriz, agua)
if contagem_lagos > 0:
    print(f'Número de lagos: {contagem_lagos}')
else: print('Não existe lagos no mapa')
```
