```python
#Questão 1

class Node:
    def __init__(self, valor, prox = None):
        self.valor = valor
        self.prox = prox
        

class Pilha:
    def __init__(self):
        self.topo = None

    def push(self, valor):
        no = Node(valor)

        if self.topo is None:
            self.topo = no
        
        else:
            atual = self.topo
            self.topo = no
            self.topo.prox = atual 

    def pop(self):
        if self.topo is None:
            return
        atual = self.topo
        self.topo = self.topo.prox
        return atual.valor

class Fila:
    def __init__(self, entrada=Pilha(), saida=Pilha()):
            self.entrada = entrada
            self.saida = saida

    def enqueue(self, valor):
        return self.entrada.push(valor)

    def dequeue(self):
        if self.saida.topo is None and self.entrada.topo is None:
            return 'Fila vazia'

        if self.saida.topo is not None:
            return self.saida.pop()
        
        while self.entrada.topo is not None:
            self.saida.push(self.entrada.pop())

        return self.saida.pop()
            
```

Explicação Q1:  
Tanto enqueue quanto dequeue são O(1). O primeiro pois é uma adição numa pilha (intuitivamente O(1)). O segundo é O(n) no pior caso, mas isso garante que todas as proximas operações de retirada serão O(1), o que faz O(n) ser amortizado para O(1).

```python
#Questão 2 e 3 e 5
#BST que eu havia previamente feito:
class Node:
    def __init__(self, valor, esquerda=None, direita=None):
        self.valor = valor
        self.esq = esquerda
        self.dir = direita
        self.nivel = None   

    def __str__(self):
        return f'Valor do nó: {self.valor}\nNível atual: {self.nivel}\nSubnó a esquerda: {self.esq.valor}\nSubnó a direita: {self.dir.valor}'


class BST:
    def __init__(self):
        self.raiz = None

    def __len__(self):
        return self._contar_nos_recursivo(self.raiz)

    def _contar_nos_recursivo(self, no_atual):
        if no_atual is None:
            return 0
        return 1 + self._contar_nos_recursivo(no_atual.esq) + self._contar_nos_recursivo(no_atual.dir)

    def get_altura(self):
        if self.raiz is None:
            return -1
        return self._get_altura_recursiva(self.raiz)

    def _get_altura_recursiva(self, no_atual):
        if no_atual is None:
            return -1
        
        altura_esq = self._get_altura_recursiva(no_atual.esq)
        altura_dir = self._get_altura_recursiva(no_atual.dir)
        
        return 1 + max(altura_esq, altura_dir)
     
    def consulta_no(self, valor):
        atual = self.raiz
        if atual is None:
            return "Árvore vazia"
        while atual is not None:
            if valor > atual.valor:
                atual = atual.dir
             
            elif valor < atual.valor:
                atual = atual.esq

            elif valor == atual.valor:
                return str(atual)
        return "Nó não encontrado"
     
    def __str__(self): #def feita por ia
        if self.raiz is None:
            return "Árvore vazia"
         
        linhas_de_texto = []
         
        self._construir_str(self.raiz, 0, linhas_de_texto)
         
        return "\n".join(linhas_de_texto) + f'\nNós na árvore: {len(self)}\nAltura da árvore: {self.get_altura()} (níveis 0 a {self.get_altura()})'

    def _construir_str(self, no_atual, nivel, linhas_de_texto):
        if no_atual is None:
            return

        self._construir_str(no_atual.dir, nivel + 1, linhas_de_texto)

        indentacao = '    ' * nivel
        linhas_de_texto.append(f"{indentacao}└── {no_atual.valor}")

        self._construir_str(no_atual.esq, nivel + 1, linhas_de_texto)




    #Algoritmo da questão 2

    def _verifica(self, atual):

        if atual is None:
            return -1
        
        altura_esq = self._verifica(atual.esq)
        if altura_esq == 'desbalanceada':
            return 'desbalanceada'
        
        altura_dir = self._verifica(atual.dir)
        if altura_dir == 'desbalanceada':
            return 'desbalanceada'
        
        if abs(altura_dir - altura_esq) > 1:
            return 'desbalanceada'
        
        return 1 + max(altura_esq, altura_dir)



    def eh_balanceada(self):
        if self._verifica(self.raiz) != 'desbalanceada':
            return 'Árvore balanceada'
        return 'Árvore desbalanceada'


    #Algoritmo questão 3:

    def _lista_recursiva(self, atual):
        if atual is None:
            return []
        
        lesq = self._lista_recursiva(atual.esq)
        ldir = self._lista_recursiva(atual.dir)

        lista_nova = lesq
        lista_nova.append(atual.valor)
        lista_nova += ldir

        return lista_nova


    def lista(self):
        return self._lista_recursiva(self.raiz)


    #Algoritmo questão 5:

    def _apara_recursivo(self, atual, high, low):
        if atual is None:
            return None
        
        atual.dir = self._apara_recursivo(atual.dir, high, low)
        atual.esq = self._apara_recursivo(atual.esq, high, low)

        if atual.valor < low:
            return atual.dir
        
        elif atual.valor > high:
            return atual.esq
        
        else:
            return atual

    def aparar(self, high, low):
        return self._apara_recursivo(self.raiz, high, low)




    def append(self, valor):
        if isinstance(valor, Node):
            no = valor
            valor = no.valor
        else:
            no = Node(valor)
         
        if self.raiz is None:
            self.raiz = no
            no.nivel = 0
            return

        atual = self.raiz
        lvl = 1
        while True:
            if valor == atual.valor:
                return KeyError("Nó já existente")
            
            if valor < atual.valor:
                if atual.esq is None:
                    atual.esq = no
                    no.nivel = lvl
                    return
                else:
                    atual = atual.esq
                    lvl += 1
            else:
                if atual.dir is None:
                    atual.dir = no
                    no.nivel = lvl
                    return
                else:
                    atual = atual.dir
                    lvl += 1
             
    def pop(self, valor=None):
        pai = None
        atual = self.raiz
        if atual is None:
            return 'Árvore vazia'
         
        if valor is None:
            if self.raiz.dir is None:
                removido = self.raiz.valor
                self.raiz = self.raiz.esq
                return f'Nó removido: {removido}'
             
            pai = self.raiz
            atual = self.raiz.dir
           
            while atual.dir is not None:
                pai = atual
                atual = atual.dir
            removido = atual.valor
            pai.dir = atual.esq

            return f'Nó removido: {removido}'
        else:
            while atual is not None and atual.valor != valor:
                pai = atual
                if valor < atual.valor:
                    atual = atual.esq
                else:
                    atual = atual.dir
                 
            if atual is None:
                return f"Valor {valor} não existe na árvore"
                 
            removido = atual
            pai_removido = pai
            nivel_removido = atual.nivel
            valor_removido = atual.valor

            if atual.esq is not None and atual.dir is not None:
                pai_substituto = atual
                substituto = atual.esq
                while substituto.dir is not None:
                    pai_substituto = substituto
                    substituto = substituto.dir
                
                if pai_substituto.esq == substituto:
                    pai_substituto.esq = substituto.esq
                else:
                    pai_substituto.dir = substituto.esq
                
                substituto.esq = removido.esq
                substituto.dir = removido.dir
                
                if pai_removido is None:
                    self.raiz = substituto
                elif pai_removido.esq == removido:
                    pai_removido.esq = substituto
                else:
                    pai_removido.dir = substituto

            elif atual.esq is not None:
                if pai_removido is None:
                    self.raiz = atual.esq
                elif pai_removido.esq == atual:
                    pai_removido.esq = atual.esq
                else:
                    pai_removido.dir = atual.esq
            else:
                if pai_removido is None:
                    self.raiz = atual.dir
                elif pai_removido.esq == atual:
                    pai_removido.esq = atual.dir
                else:
                    pai_removido.dir = atual.dir
                         
            return f'Nó removido: {valor_removido}; No nível: {nivel_removido}'
        


tree = BST()

while True:
    N = int(input("Qual operação deseja realizar?\n 1- Append;  2- Pop; 3- Consultar nó; 4- Verificar se é balanceada; 5- Representação em lista; 6- Aparar árvore; 0- Parar\n"))
    
    if N == 0:
        break

    elif N == 1:
        v = float(input('Valor do nó a ser adicionado:\n'))
        tree.append(v)
        print()
        print(tree)
        print()

    elif N == 2:
        v = input('Valor do nó a ser removido: (vazio caso queira o último)\n')
        if v == '':
            tree.pop()
        else:
            v = float(v)
            tree.pop(v)

        print()
        print(tree)
        print()

    elif N == 3:
        v = float(input('Valor do nó a ser consultado:\n'))
        print(tree.consulta_no(v))
        print()

    elif N == 4:
        print(tree.eh_balanceada())

    elif N == 5:
        print(tree.lista())

    elif N == 6:
        low = float(input('Qual o parâmetro low?\n'))
        high = float(input('Qual o parâmetro high\n'))
        tree.aparar(high, low)
        print(tree)
    else:
        print('Valor incopatível, tente novamente.')
```
